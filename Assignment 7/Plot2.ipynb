```python
import matplotlib.pyplot as plt
import math
#By putting the initial values we obtain c1=1 and c2=1
def f7(x):
    return 1 + pow(math.e,-x) - (x**2)/2 + 2*x
xval = []
yval = []
#taking a step size of 0.05, there are 10/0.05= 200 iterations
for i in range(200):
    xval.append(-5+ i*0.05)
    yval.append(f7(xval[i]))
plt.plot(xval,yval)
plt.xlabel("x")
plt.ylabel("y")
plt.title("y v x")
```
