```python
import matplotlib.pyplot as plt
import math
def factorial(n):                              #Factorial as Taylor expasnion
    fact =1
    for i in range(1,n+1):
        fact = fact*i
    return fact
def sine(x,n):                                #sine function as Taylor expansion
    n = math.floor(n)
    s = 0
    for i in range(n):
        s = s+ pow(-1,i)*pow(x,2*i+1)/factorial(2*i+1)
    return s
def sine_error(x,n):                       #Error in sine
    error = abs(math.sin(x)-sine(x,n)) 
    return error
i = 1
iter = []             #to store the iteration numbers
err = []              #to store error
while sine_error(0.08, i)>0.000000000001: 
    iter.append(i)
    err.append(sine_error(0.08,i))
    i = i+1
plt.plot(iter, err,".-")
plt.xlabel("Iteration Number")
plt.ylabel("Modulus of Error in Sine")
plt.title("Modulus of error in sine vs iteration number")
plt.show()
```
