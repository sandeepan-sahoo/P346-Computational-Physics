```python
import matplotlib.pyplot as plt
import math
def factorial(n):
    fact = 1
    for i in range(1,n+1):
        fact = fact*i
    return fact
def exp(x,n):
    n = math.floor(n)
    s = 0
    for i in range(n):
        s = s+ pow(x,i)/factorial(i)
    return s
def exp_error(x,n):
    error = abs(math.exp(x)-exp(x,n))
    return error
i = 1
iter = []
err = []
while exp_error(0.08, i)>0.000000000001:
    iter.append(i)
    err.append(exp_error(0.08,i))
    i = i+1
plt.plot(iter, err)
plt.xlabel("Iteration Number")
plt.ylabel("Modulus of error in exponential")
plt.title("Error in exponential vs iteration number")
plt.show()
```

```python
import matplotlib.pyplot as plt
import math
def factorial(n):
    fact =1
    for i in range(1,n+1):
        fact = fact*i
    return fact
def sine(x,n):
    n = math.floor(n)
    s = 0
    for i in range(n):
        s = s+ pow(-1,i)*pow(x,2*i+1)/factorial(2*i+1)
    return s
def sine_error(x,n):
    error = abs(math.sin(x)-sine(x,n))
    return error
i = 1
iter = []
err = []
while sine_error(0.08, i)>0.000000000000001:
    iter.append(i)
    err.append(sine_error(0.08,i))
    i = i+1
plt.plot(iter, err, ".-")
plt.xlabel("Iteration Number")
plt.ylabel("Modulus of Error in Exponential")
plt.title("Modulus of error in exp vs iteration number")
plt.show()
```
